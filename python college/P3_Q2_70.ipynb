#### 5. Write a Python Program to Count the Number of Each Vowel, consonants and spaces in the given String(Multiline) using Dictionary (Use V_dict for vowels and C_dict for consonants and spaces)<a href="#5.-Write-a-Python-Program-to-Count-the-Number-of-Each-Vowel,-consonants-and-spaces-in-the-given-String(Multiline)-using-Dictionary-(Use-V_dict-for-vowels-and-C_dict-for-consonants-and-spaces)" class="anchor-link">¶</a>

In \[5\]:

    d={"vowel_count":0,"character_count":0,"space_count":0}
    s='''My name is Yashasvi Srivastava'''
    l=list(s)

    vowels=['a','e','i','o','u']
    for i in l:
        if i!='\n':
            if i in vowels:
                d['vowel_count']+=1
            elif i==" ":
                d['space_count']+=1
            else:
                d['character_count']+=1
    print(d)

    {'vowel_count': 10, 'character_count': 16, 'space_count': 4}
